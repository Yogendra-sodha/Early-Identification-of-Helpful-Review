- Removed ASCII Characters from the reviews

```python
def filterSpecialChars(strInput):
    result = []
    for character in (str(strInput)):
        ordVal = ord(str(character))
        if ordVal < 0 or ordVal > 127:
            result.append(' ')
        else:
            result.append(str(character))
    return ''.join(result)
```

```python
df["Review"] = df["Review"].apply(filterSpecialChars)
```

#### Worcloud after remove ASCII characters

#### Expanding Feature Spcae
- Adding helpful percent which is likes on reviews divided by total votes where total votes is like and dislikes on the reviews
- Added Review length of every review

```python
like=df["Like"]
dis=df["Dislike"]
df['Total_votes'] = (like) + dis
df['helpful_percent'] = df.apply(lambda row: row['Like']/row['Total_votes'] 
                                 if row['Total_votes'] else 0, axis=1)
df['Review_Length'] = df['Review'].apply(lambda x: len([y for y in (str(x)).split()]))
```

#### Using Vader library as taught in lecture
- Adding the sentiment of all reviews
- Extracting all the sentiments from sentiment_score and creating specific column for particular sentiment

```python
sid = SentimentIntensityAnalyzer()
df["Sentiment_Score"] = df["Review"].apply(lambda x : sid.polarity_scores(str(x)))
print(df["Sentiment_Score"])
df["Negative"] = df["Sentiment_Score"].apply( lambda x: x['neg'])
df["Neutral"] = df["Sentiment_Score"].apply( lambda x : x['neu'])
df["Positive"] = df["Sentiment_Score"].apply(lambda x : x ['pos'])
df["Compound"] = df["Sentiment_Score"].apply(lambda x: x['compound'])
```

```python
df.head(4)
```

```python
def getAnalysis(score):
    if score > 0.75:
        return 1
    elif score == 0:
         return 0
    else:
        return 0
df["helpful"] = df["helpful_percent"].apply(getAnalysis)
df.index = [x for x in range(1, len(df.values)+1)]
```

#### Scaling

```python
scaler = MinMaxScaler()
```

```python
df[['scaled_star_rating','scaled_review_Length']] = scaler.fit_transform(df[['Star','Review_Length']])
df = df[df.Total_votes >3 ]
```
