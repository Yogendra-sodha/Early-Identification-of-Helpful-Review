### Scrapping the reviews using Selenium
- The function will take link of review as parameter and type of driver as parameter.
- The function will return dataframe with columns Title,Reviewer_name,Rating,Product,Date,Review,Likes, Dislike, total_person scraped from the function.

```python
import requests
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from bs4 import BeautifulSoup 
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions 

def get_N_Reviews(page_url, driver):
    
    reviews = []
    title =[]
    reviewer_name=[]
    date =[]
    rating=[]
    review_content=[]
    helpful =[]
    non_helpful=[]
    for i in range(1,10):
        page_url=page_url+"&page="+str(i)
        page=requests.get(page_url)
        driver.get(page_url)
        driver.implicitly_wait(2)
        #more_reviews= driver.find_element(By.CSS_SELECTOR,"body div#container div._2tsNFb div._1YokD2._3Mn1Gg.col-9-12 div._1AtVbE.col-12-12 div._2MImiq._1Qnn1K nav.yFHi8N a._1LKTO3")   
        #more_data=driver.find_element(By.CSS_SELECTOR,"span._1BWGvX span")
        content = driver.page_source
        soup = BeautifulSoup(content, 'html.parser')
        show_title=soup.select("body div#container div div._2tsNFb div._6t1WkM._3HqJxg div._1YokD2._2GoDe3.col-12-12 div._1AtVbE.col-10-12 div._2s4DIt._1CDdy2")[0].get_text().strip()
        divs=soup.select("body div#container div div._2tsNFb div._6t1WkM._3HqJxg div._1YokD2._2GoDe3.col-12-12 div._1YokD2._3Mn1Gg.col-9-12 div._1AtVbE.col-12-12 div._27M-vq div.col._2wzgFH.K0kLPL ")
        for idx,div in enumerate(divs):
            overall= div.select("div.row.Bfvo5l div.col-4-12._17ETNY div.col-4-12._17ETNY div._2d4LTz")
            #div_show=overall_title.select("")
    
            div_rating=div.select("div.col._2wzgFH.K0kLPL div._3LWZlK._1BLPMq")
   
            div_title= div.select("div.row p._2-N8zT ")

            div_reviewer_name=div.select("div._27M-vq div.col._2wzgFH.K0kLPL div.row._3n8db9 div.row p._2sc7ZR._2V5EHH")

            div_date =div.select("div._27M-vq div.col._2wzgFH.K0kLPL div.row._3n8db9 div.row p[class=_2sc7ZR]")

            div_review_short=div.select(" div.col._2wzgFH.K0kLPL div._3LWZlK._1BLPMq p._2-N8zT ")
            div_review_content=div.select("div._27M-vq div.col._2wzgFH.K0kLPL div.row div.t-ZTKy div")
    
            div_helpful =div.select(" div._27M-vq div.col._2wzgFH.K0kLPL div.row._3n8db9 div._1e9_Zu div._27aTsS div._1LmwT9 span._3c3Px5 ")
            div_non=div.select("div._27M-vq div.col._2wzgFH.K0kLPL div.row._3n8db9 div._1e9_Zu div._27aTsS div._1LmwT9.pkR4jH span._3c3Px5")
            more=div.select("div._27M-vq div.col._2wzgFH.K0kLPL div.row div.t-ZTKy span._1BWGvX")
            if div_rating!=[]:
                rating_one=div_rating[0].get_text()
            else: 
                rating_one="0"
            #print(rating_one)
            if div_title!=[]:
                title_one=div_title[0].get_text()  
            else: 
                title_one="no_title"
            #print(title_one)
            if div_reviewer_name!=0:
                name_one=div_reviewer_name[0].get_text()
            else: 
                name_one="aaa"
            #print(name_one)
            if div_date!=[]:
                date_one=div_date[0].get_text()
            else: 
                date_one="Nov 2021"
            #print(date_one)
            if div_review_content!=[]:
                sam_content=div_review_content[0].get_text().strip().split("READ MORE")
                content=" ".join(sam_content)
                
            else: 
                content="no content"
            #print(content)
            if div_helpful!=[]:
                help_one=div_helpful[0].get_text()
                help_num=int(help_one)
            else: 
                help_one="Non_help"
            #print(help_one)
            if div_non!=[]:
                non_help_one=div_non[0].get_text()
                non_help_num=int(non_help_one)
            else: 
                non_help_one=0
            total_person=help_num+non_help_num
            
            #print(non_help_one)
            #print(total_person)
            reviews.append((title_one,name_one,rating_one,show_title,date_one,content,help_one,non_help_one,total_person))
        df= pd.DataFrame(reviews,columns=['Title', 'Reviewer_name', 'Rating',"Product",'Date', 'Review',
                                               'Likes', 'Dislike', 'total_person'])
    return df
```
